#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
    Dica:
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis,
        3. conte o número de categorias dessa variável
        4. acumule essa informação de 3. na lista que você criou em 1.
        5. No final, essa lista pode ser atribuída à nossa variável.

````python
nomes_variaveis = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'qtd_filhos',
       'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia', 'idade',
       'tempo_emprego', 'possui_celular', 'possui_fone_comercial',
       'possui_fone', 'possui_email', 'qt_pessoas_residencia', 'mau']

# Lista com os tipos de dados correspondentes
tipos_dados = ['object', 'object', 'object', 'int64', 'object', 'object', 'object', 'object', 'float64',
              'float64', 'int64', 'int64', 'int64', 'int64', 'float64', 'bool']

# Criando um dicionário com os dados
dados_metadados = {'Nome da Variável': nomes_variaveis, 'Tipo de Dado': tipos_dados}

# Criando o DataFrame de metadados
df_metadados = pd.DataFrame(dados_metadados)

print(df_metadados)
````

````python
# Adicionando a coluna 'Descrição'
dados_metadados['Descrição'] = ['Sexo da pessoa', 'Possui veículo próprio', 'Possui imóvel próprio', 'Número de filhos',
                               'Tipo de renda principal', 'Nível de escolaridade', 'Estado civil', 'Tipo de residência',
                               'Idade em anos', 'Tempo de trabalho no emprego atual', 'Possui celular',
                               'Possui telefone comercial', 'Possui telefone residencial', 'Possui email',
                               'Número de pessoas que residem na mesma casa', 'Indicador de inadimplência']

# Criando o DataFrame de metadados
df_metadados = pd.DataFrame(dados_metadados)
````

````python
def adicionar_qtd_categorias(df_metadados):

  # Lista para armazenar a quantidade de categorias
  qtd_categorias = []

  # Iterando sobre as variáveis
  for variavel in df_metadados['Nome da Variável']:
    # Obtendo a série correspondente à variável no DataFrame original (substitua 'df' pelo nome do seu DataFrame)
    serie = df[variavel]
    # Contando o número de categorias únicas e adicionando à lista
    qtd_categorias.append(serie.nunique())

  # Adicionando a coluna ao DataFrame de metadados
  df_metadados['qtd_categorias'] = qtd_categorias

  return df_metadados

# Exemplo de uso:
df_metadados = adicionar_qtd_categorias(df_metadados)
df_metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# Identificando as colunas categóricas
colunas_categoricas = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']

# Criando um novo DataFrame com as variáveis dummy
df_dummy = pd.get_dummies(df[colunas_categoricas], drop_first=True)

# Concatenando o DataFrame original com o DataFrame de variáveis dummy (excluindo as colunas categóricas originais)
df_final = pd.concat([df.drop(colunas_categoricas, axis=1), df_dummy], axis=1)

df_final
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
tabela_cruzada = pd.crosstab(df['possui_email'], df['posse_de_veiculo'], values=df['mau'], aggfunc='mean')

print(tabela_cruzada)
````

A váriavel mais poderosa para prever a probabilidade de mau = 1 é a "posse_de_veiculo".
