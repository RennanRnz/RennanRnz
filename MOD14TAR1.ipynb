- ## Crie um jupyter notebook com os imports necessários e um código para gerar o gráficos referentes a base de dados de março, abril, maio, junho e dezembro (numa célula só ou em várias).


```python
# Importando as bibliotecas que iremos utilizar.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SINASC_RO_2019.csv')
df['DTNASC'] = pd.to_datetime(df['DTNASC'])

# Convertendo a coluna de data de nascimento para o tipo datetime e extraindo o mês.
df['DTNASC'] = pd.to_datetime(df['DTNASC'])
df['DTNASC'] = df['DTNASC'].dt.month

# Meses desejados.
meses = [3, 4, 5, 6, 12]
df = df[df['DTNASC'].isin(meses)]
```

```python
for mes in meses:
    df_mes = df[df['DTNASC'] == mes]

    # Agrupando por APGAR5 e contando as ocorrências.
    df_mes_agrupado = df_mes.groupby('APGAR5').size().reset_index(name='contagem')

    # Criando o gráfico de linha
    plt.figure(figsize=(10, 6))
    plt.plot(df_mes_agrupado['APGAR5'], df_mes_agrupado['contagem'])
    plt.title(f'Gráfico para o mês {mes}')
    plt.xlabel('Valor de APGAR5')
    plt.ylabel('Frequência')
    plt.grid(True)
    plt.show()
```

- ## Como que você poderia fazer para deixar isso mais automatizado? Descreva na última célula do jupyter notebook o que ajudaria na hora de gerar novos gráficos pros meses/anos seguintes.

```python
# Função para automatizar o processo
def gerar_graficos(arquivo, meses):
    df = pd.read_csv(arquivo)
    df['DTNASC'] = pd.to_datetime(df['DTNASC'])
    df['DTNASC'] = df['DTNASC'].dt.month
    df = df[df['DTNASC'].isin(meses)]
    for mes in meses:
        df_mes = df[df['DTNASC'] == mes]
        plt.figure(figsize=(10, 6))
        plt.plot(df_mes_agrupado['APGAR5'], df_mes_agrupado['contagem'])
        plt.title(f'Gráfico para o mês {mes}')
        plt.xlabel('Valor de APGAR5')
        plt.ylabel('Frequência')
        plt.grid(True)
        plt.show()

# Agora podemos utilizar diferentes parâmetros.
gerar_graficos('SINASC_RO_2019.csv', [3, 4, 5, 6, 12])
```
