#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sinasc_RO_2019.csv.csv', na_values='na')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Gráfico de barras para APGAR1
sns.countplot(x='APGAR1', data=df)
plt.title('Frequência dos Escores APGAR1')
plt.xlabel('Escore APGAR1')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barras para APGAR5
sns.countplot(x='APGAR5', data=df)
plt.title('Frequência dos Escores APGAR5')
plt.xlabel('Escore APGAR5')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barras para SEXO
sns.countplot(x='SEXO', data=df)
plt.title('Frequência por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barras para GESTACAO
sns.countplot(x='GESTACAO', data=df)
plt.title('Frequência por Duração da Gestação')
plt.xlabel('Semanas de Gestação')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barras para ESTCIVMAE
sns.countplot(x='ESTCIVMAE', data=df)
plt.title('Frequência por Estado Civil da Mãe')
plt.xlabel('Estado Civil da Mãe')
plt.ylabel('Frequência')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhor visualização
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Agrupando os dados por consulta e calculando o APGAR5 médio
df_agrupado = df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()

# Criando o gráfico de barras
sns.barplot(x='CONSULTAS', y='APGAR5', data=df_agrupado)
plt.xlabel('Número de CONSULTAS')
plt.ylabel('APGAR5 Médio')
plt.title('Evolução do APGAR5 Médio por Consulta')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
# Criando um subplot com 2 linhas e 2 colunas
fig, axes = plt.subplots(2, 2, figsize=(8, 10))

# Gráfico 1: Frequência por Sexo
sns.countplot(x='SEXO', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Frequência por Sexo')

# Gráfico 2: Frequência por Gestação
sns.countplot(x='GESTACAO', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Frequência por Gestação')

# Gráfico 3: Frequência por Estado Civil da Mãe
sns.countplot(x='ESTCIVMAE', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Frequência por Estado Civil da Mãe')

# Gráfico 4: Frequência por Número de Consultas
sns.countplot(x='CONSULTAS', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Frequência por Número de Consultas')

# Ajustando o layout
plt.tight_layout()
plt.show()
```
